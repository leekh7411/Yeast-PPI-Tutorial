## Dataset for Visualization
- 시각화는 그 목적 자체가 전체 데이터셋을 관찰하기 위함 보다는 사람이 알아 볼 수 있고 스스로 분석할 수 있을 정도의 볼륨을 유지해야한다
- 따라서 이를 위한 추상화 작업 혹은 일부만 샘플링해서 사용
- 아래 예제 코드에서는 무작위로 edge를 샘플링하여 사용하고 있음
- 만약 무작위 샘플링이 아니라 특정 규칙으로 샘플링을 하면 새로운 시각화 결과를 관찰 가능 

```python
# Load Yeast Protein Protein Interaction dataset
# - Exists as edge list
import numpy as np
import random

def load_edge_list(path, num_edges=2500):
    edge_list = open(path, 'r')
    edge_list = edge_list.readlines()
    edges = []
    nodes = []
    edge_list = random.sample(edge_list, num_edges)
    for edge in edge_list:
        edge = edge.replace("\n", "")
        edge_vals = edge.split("\t")
        edges.append((edge_vals[0], edge_vals[1]))
        nodes.append(edge_vals[0])
        nodes.append(edge_vals[1])
    
    nodes = list(set(nodes)) # Remove redundant values -> set object to list object
    return edges, nodes
```

```python
edge_path = "yeast.reduced.edgelist"
edge_list, node_list = load_edge_list(edge_path)

# For annotation..
print("Number of edges : {}".format(len(edge_list)))
for i, edge in enumerate(edge_list):
    if i > 10: break
    print("Protein ID {:>10} is connected with Protein ID {:>10} in the Yeast PPI Graph.".format(edge[0],edge[1]))
```

```python
# How to visualize the target PPI Graph?
# - Use NetworkX (Python Lib)
# - documents : https://networkx.github.io/documentation/stable/
import networkx as nx

# Make networkx graph object directly from edge list
raw_edge_list = open(edge_path, "r")
G = nx.Graph()
G = nx.parse_edgelist(raw_edge_list, delimiter='\t', create_using=G,nodetype=str, data=(('weight', float),))

# Or add edges and nodes from lists to networkx graph object
G = nx.Graph()
G.add_nodes_from(node_list)
G.add_edges_from(edge_list)

# Get graph edges and nodes from networkx graph object
nodes = G.nodes
edges = G.edges

print("Graph's Nodes : {} / Edges : {}".format(len(nodes), len(edges)))
```

## Example 1
- 무작위로 샘플링 된 edge를 바탕으로 구성되는 여러 그래프를 시각화

```python
# How to Visualize Graph?
import numpy as np
import matplotlib.pyplot as plt

title = "Yeast-PPI-Network-Full-Dataset"
plt.figure(figsize=(10,10))
node_size = 15
node_alpha = .6

edge_color = '#483D8B'
edge_alpha = 0.15
edge_style = 'dashed'
edge_width = 2

protein_node_color = "#F08080"
rna_node_color = "#1E90FF"

pos = nx.drawing.nx_agraph.graphviz_layout(G)

nx.draw_networkx_nodes(G, pos, node_size=node_size, 
                       node_color=rna_node_color, 
                       alpha=node_alpha)

nx.draw_networkx_edges(G, pos, width=edge_width, 
                       alpha=edge_alpha)

plt.axis('off')
plt.savefig("{}.png".format(title))
plt.title("{} / Blue - Protein".format(title))
plt.show()
```

## Example 2
- 무작위로 샘플링 된 edge를 바탕으로 구성되는 여러 그래프를 시각화
- 그래프를 그릴때 node의 위치를 결정하는 알고리즘을 networkx에서는 다양하게 제공하고 있음

```python
# How to Visualize Graph?
import numpy as np
import matplotlib.pyplot as plt

title = "Yeast-PPI-Network-Full-Dataset"
plt.figure(figsize=(10,10))
node_size = 15
node_alpha = .6

edge_color = '#483D8B'
edge_alpha = 0.15
edge_style = 'dashed'
edge_width = 2

protein_node_color = "#F08080"
rna_node_color = "#1E90FF"

# 연결성이 끊어진, 작은 다수의 그래프를 시각화 할 때 사용하기 좋은 layout
#pos = nx.drawing.nx_agraph.graphviz_layout(G) 

# Node의 위치를 무작위로 결정함
# 장점은 위치 결정 알고리즘을 사용하지 않기 때문에 연산 시간이 빠름
# 단점은 빠르기만하고 시각화 했을 경우 알아보기 힘들다 
#pos = nx.random_layout(G)

# Kamada 라는 연구자가 발명한 그래프 레이아웃 알고리즘
# 스프링 시스템의 에너지를 최소화 하는 방향으로 그래프를 그리는 알고리즘
#pos = nx.kamada_kawai_layout(G)

# Circular layout 
# 모든 node를 원형 위치로 구성해 그래프를 그린다
pos = nx.circular_layout(G)

nx.draw_networkx_nodes(G, pos, node_size=node_size, 
                       node_color=rna_node_color, 
                       alpha=node_alpha)

nx.draw_networkx_edges(G, pos, width=edge_width, 
                       alpha=edge_alpha)

plt.axis('off')
plt.savefig("{}.png".format(title))
plt.title("{} / Blue - Protein".format(title))
plt.show()
```

## Example 3
- 무작위로 샘플링 된 edge를 바탕으로 구성되는 여러 그래프를 시각화
- 그래프를 그릴때 node의 위치를 결정하는 알고리즘을 networkx에서는 다양하게 제공하고 있음
- 그래프의 노드 크기, 색상, 엣지의 색상, 크기, 선 스타일, 두께, 투명도 등 다양하게 제어 가능

```python
# How to Visualize Graph?
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

edge_path = "yeast.reduced.edgelist"
edge_list, node_list = load_edge_list(edge_path, num_edges=50)
G = nx.Graph()
G.add_nodes_from(node_list)
G.add_edges_from(edge_list)

node_list_A = node_list[:len(node_list)//2]
node_list_B = node_list[len(node_list)//2:]

edge_list_A = edge_list[:len(edge_list)//2]
edge_list_B = edge_list[len(edge_list)//2:]

title = "Yeast-PPI-Network-Full-Dataset"
plt.figure(figsize=(10,10))
node_size = 15
node_alpha = .6

edge_color = '#483D8B'
edge_alpha = 0.15
edge_style = 'dashed'
edge_width = 2

blue_node_color = "#F08080"
pink_node_color = "#1E90FF"

# 연결성이 끊어진, 작은 다수의 그래프를 시각화 할 때 사용하기 좋은 layout
pos = nx.drawing.nx_agraph.graphviz_layout(G) 

# Node의 위치를 무작위로 결정함
# 장점은 위치 결정 알고리즘을 사용하지 않기 때문에 연산 시간이 빠름
# 단점은 빠르기만하고 시각화 했을 경우 알아보기 힘들다 
#pos = nx.random_layout(G)

# Kamada 라는 연구자가 발명한 그래프 레이아웃 알고리즘
# 스프링 시스템의 에너지를 최소화 하는 방향으로 그래프를 그리는 알고리즘
#pos = nx.kamada_kawai_layout(G)

# Circular layout 
# 모든 node를 원형 위치로 구성해 그래프를 그린다
#pos = nx.circular_layout(G)

nx.draw_networkx_nodes(G, pos, 
                       node_size=100, # 그래프에서 그려지는 노드의 크기를 결정
                       node_color=blue_node_color, # 그래프에서 그려지는 노드의 색상을 결정
                       nodelist = node_list_A, # 그래프 객체 G에 넣은 node 정보 중 원하는 node의 리스트를 별도로 주면 해당 node만 색칠한다
                       alpha=0.9) # 그래프에서 그려지는 노드의 투명도를 결정 (0 ~ 1)

nx.draw_networkx_nodes(G, pos, 
                       node_size=100, # 그래프에서 그려지는 노드의 크기를 결정
                       node_color=pink_node_color, # 그래프에서 그려지는 노드의 색상을 결정
                       nodelist = node_list_B, # 그래프 객체 G에 넣은 node 정보 중 원하는 node의 리스트를 별도로 주면 해당 node만 색칠한다
                       alpha=0.9) # 그래프에서 그려지는 노드의 투명도를 결정 (0 ~ 1)

nx.draw_networkx_edges(G, pos, 
                       style="dashed", # 그래프에서 그려지는 엣지의 스타일을 결정 (solid|dashed|dotted,dashdot)
                       width=4, # 그래프에서 그려지는 엣지의 두께를 결정
                       edge_color = "green", # 그래프에서 그려지는 엣지의 색상을 결정
                       edgelist= edge_list_A,
                       alpha=1)

nx.draw_networkx_edges(G, pos, 
                       style="dashed", # 그래프에서 그려지는 엣지의 스타일을 결정 (solid|dashed|dotted,dashdot)
                       width=4, # 그래프에서 그려지는 엣지의 두께를 결정
                       edge_color = "yellow", # 그래프에서 그려지는 엣지의 색상을 결정
                       edgelist= edge_list_B,
                       alpha=1)

plt.axis('off')
plt.savefig("{}.png".format(title))
plt.title("{}".format(title))
plt.show()
```
